Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
```

Data collection and Pre-processing

```python
#loading a dataset from a csv file to a pandas dataframe
raw_text_data = pd.read_csv('dataset.csv')
```

```python
print(raw_text_data)
```

```python
#replacing the null values with a null string
text_data=raw_text_data.where((pd.notnull(raw_text_data)),'')
```

```python
#printing the first 5 rows of the dataframe
text_data.head()
```

```python
#checking the number of rows and columns in the dataframe
text_data.shape
```

Seperating data as text and labels

```python
X=text_data['text']
Y=text_data['1']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature extraction(converting of text data to numerical values ,i.e feature vectors)

```python
#transforming the text data into feature vectors that can be used as input to the logistic regression model
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)  # Change 'True' to True
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
#convert X_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

Training the model

Logistic Regression



```python
model=LogisticRegression()
```

```python
#training the logistic regression model using training data
model.fit(X_train_features,Y_train)
```

Evaluating the trained model

```python
#prediction on training data
prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
```

```python
print('Accuracy on training data :',accuracy_on_training_data)
```

```python
#prediction on test data
prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
```

```python
print('Accuracy on test data :',accuracy_on_test_data)
```

Building a predictive system

```python
input_text=["my credit card number is 4111 1111 1111 1111"]
input_data_features=feature_extraction.transform(input_text)
prediction=model.predict(input_data_features)
print(prediction)
if (prediction[0]==1):
  print('Sensitive information is detected')
else:
  print('No sensitive information is detected')
```
